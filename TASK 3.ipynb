**VISUALIZATION OF DATASET**

```python
# Visualization of Dataset
import seaborn as sns
import matplotlib.pyplot as plt
```

**SET OBJECTIVES FOE VISUALIZATION**

Ask Meaningful Questions like:

```python
# 1. How is the gender of students distributed in dataset?
sns.countplot(x=df['Gender'], palette='Blues')
plt.title('Gender Distribution')
plt.figure(figsize=(3,5) )
plt.show()
```

**Interpretation:**
**Gender Distribution( Bar Chart):**
 Shows how many females and males  students are in dataset.

```python
# 2. What is the  overall  distribution  of Social media addiction  scores among students?
sns.histplot(data= df ,x='Addicted_Score'  ,  kde =True  ,  bins=10 , color='purple' )
plt.title('Addicted Score Distribution')
plt.show()
```

**Interpretation:**
**Distribution  of Addicted Scores (Histogram):**
Most students have an addiction score between  **5 and 9**, with very few scoring below 3 .A majority of students demonstrate **moderate to high levels**  of social media addiction.

```python
# 3. Which social media platform are  most popular ?
sns.countplot(df['Most_Used_Platform'], palette='Set1')
plt.title('Most Used Social Media Platform')
plt.show()
```

**Interpretation:**
**Most Used Social Media Platforms**:
Bar chart shows  that Instagram and TikTok    are the most popular platforms  among  students.

```python
# 4. Do addiction score different between male and female students?
sns.boxplot(data= df, x='Gender' , y='Addicted_Score' , palette='Set2')
plt.title('Addicted Score by Gender')
plt.show()
```

**Interpretation:**
**Addicted_Score by Gender (Boxplot)**:


*   Median addicted score is  slightly higher in females than males ,so Social Media addiction is  more intense among female student.

*   Both  genders show a wide  range  of scores and contain outliers .


```python
# 5. Is there a relationship between time spent on social media and sleep duration?
sns.set(style = "whitegrid")
sns.scatterplot(data = df , x='Avg_Daily_Usage_Hours'  , y= 'Sleep_Hours_Per_Night'  , hue = 'Gender' , palette='coolwarm', alpha = 0.7)
plt.title('Daily Use of Social Media vs Sleep  Hours')
plt.xlabel('Average Daily Usage Hours')
plt.ylabel('Sleep Hours Per Night ')
plt.show()
```

**Interpretation:**
**Avg_Daily_Usage vs Sleep_Hours (Scatter plot ):**


*   The scatter plot shows a slight  downward trend , suggesting that as average daily social medial platform increases , sleep hours tend to decrease.

*  This trend appears across both males and females .

*   Females show more data points with higher daily usage and slightly reduced sleep compared to males





```python
# 6. Which numeric varaibles are strongly correlated with social media addicted score?
corr_matrix=df.corr(numeric_only=True)
sns.heatmap(corr_matrix ,annot=True , cmap='coolwarm')
plt.title('Correlation Heatmap of Student Social Media Data')
plt.show()
```

**Interpretation**:
**Correlation between Addicted_Score and other Variables:**


*   Addicted_Score  has  a strong positive  correlation with:
             

         Avg_Daily_Usage_Hours(more usage =higher addiction )
         Conflict_Over_Social_Media


*  Week Correlation with  Sleep Hours and Mental Health Score
